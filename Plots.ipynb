# Activity vs time

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

isotopes = ["11-C", "15-O", "12-N", "10-C", "14-O", "8-B"]
rates = [4.6e-3, 2.02e-3, 2.4e-4, 1.4e-4, 3e-5, 3.2e-5] # 55 MeV values from Dendooven
#rates = [1.33e-2, 7.22e-3, 2.33e-3, 4.51e-4, 4.16e-4, 1.12e-5] # 120 MeV values
T = [20.36*60, 2.04*60, 11e-3, 19.29, 1.17*60, 770e-3]
lamb = np.log(2)/T

def activity(i, t):
    return rates[i]*(1-np.exp(-lamb[i]*t))

def totnumber(i, t):
    return rates[i]*(t + 1/lamb[i]*(np.exp(-lamb[i]*t)-1))

ts = np.linspace(0, 3, 500)
plt.figure()
fig0, ax0 = plt.subplots(1, 2, figsize=(10,3.8))
activity0 = activity(0, ts)
activity1 = activity(1, ts)
activity2 = activity(2, ts)
activity3 = activity(3, ts)
activity4 = activity(4, ts)
activity5 = activity(5, ts)
ax0[0].plot(ts, activity0, label=isotopes[0])
ax0[0].plot(ts, activity1, label=isotopes[1])
ax0[0].plot(ts, activity2, label=isotopes[2])
ax0[0].plot(ts, activity3, label=isotopes[3])
ax0[0].plot(ts, activity4, label=isotopes[4])
ax0[0].plot(ts, activity5, label=isotopes[5])
ax0[0].legend()
ax0[0].grid()

ts = np.linspace(0, 240, 5000)
plt.figure()
fig, ax = plt.subplots(1, 2, figsize=(10,3.8))
tot0 = totnumber(0, ts)
tot1 = totnumber(1, ts)
tot2 = totnumber(2, ts)
tot3 = totnumber(3, ts)
tot4 = totnumber(4, ts)
tot5 = totnumber(5, ts)
ax[0].plot(ts, tot1, label=isotopes[1], color='green', linestyle="-")
ax[0].plot(ts, tot2, label=isotopes[2], color='red', linestyle="--")
ax[0].plot(ts, tot0, label=isotopes[0], color='blue', linestyle="-.")
ax[0].plot(ts, tot3, label=isotopes[3], color='magenta', linestyle=":")
ax[0].plot(ts, tot5, label=isotopes[5], color='darkgoldenrod', linestyle=":")
ax[0].plot(ts, tot4, label=isotopes[4], color='black', linestyle=":")
ax[0].legend(fontsize=11.5)
#plt.yscale('log')
#plt.ylim(1e-9,1e-2)
ax[0].grid()
ax[0].set_ylabel('integrated number of decays', fontsize=13)
ax[0].set_xlabel('time [s]', fontsize=13)
ax[0].text(90, 0.18, 'Proton 55 MeV', fontsize=13)
ax[0].text(90, 0.155, 'PMMA target', fontsize=13)
ax[0].tick_params(labelsize=13)

ax[1].plot(ts, tot2, label=isotopes[2], color='red', linestyle="--")
ax[1].plot(ts, tot1, label=isotopes[1], color='green', linestyle="-")
ax[1].plot(ts, tot3, label=isotopes[3], color='magenta', linestyle=":")
ax[1].plot(ts, tot0, label=isotopes[0], color='blue', linestyle="-.")
ax[1].plot(ts, tot5, label=isotopes[5], color='darkgoldenrod', linestyle=":")
ax[1].plot(ts, tot4, label=isotopes[4], color='black', linestyle=":")
ax[1].legend(fontsize=11.5, ncol=3)
plt.yscale('log')
plt.ylim(1e-9,1e-2)
#plt.ylim(2e-9,0.003)
plt.xlim(0, 10)
ax[1].grid()
ax[1].set_xlabel('time [s]')
ax[1].text(5.8, 1.4e-6, 'Proton 55 MeV', fontsize=13)
ax[1].text(5.8, 2e-7, 'PMMA target', fontsize=13)
ax[1].tick_params(labelsize=13)
fig.savefig("nodecays.pdf", dpi=150)
```
